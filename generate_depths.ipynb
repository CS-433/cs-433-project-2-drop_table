```python
def extract_zip(input_zip):
    input_zip=zipfile.ZipFile(input_zip)
    return {name: input_zip.read(name) for name in input_zip.namelist()}
```

```python
import numpy as np
import ast
import math
from haversine import haversine_vector, Unit
import numpngw
```

```python
def file_to_py(path):
    with open(path,'r') as file:
        myfile = file.read()
        return txt_to_py(myfile)
```

```python
def txt_to_py(txt):
    index = txt.find(']')
    drone = ast.literal_eval(txt[1:index+1])
    pixels = ast.literal_eval("["+ txt[index+2:])
    return drone, pixels
```

```python
def compute_depth(drone, pixels):
    l = pixels[-1]
    c,r = l[1]

    y_drone,x_drone,z_drone,_,_,_ = drone[0]
    drone_coord = [(x_drone,y_drone)]
  
    pixels_coord = [(x,y) for [(x,y,_),(_,_)] in pixels]
    pixels_invalid = [(x==-1) for [(x,_,_),(_,_)] in pixels]
    pixels_alt = np.array([z for [(_,_,z),(_,_)] in pixels])
  
    haversine_distance = np.squeeze(haversine_vector(drone_coord, pixels_coord, Unit.METERS, comb=True))
    distance = np.sqrt(haversine_distance**2+(pixels_alt-z_drone)**2)
    distance[np.where(pixels_invalid)] = 0
  
    matrix_distance = distance.reshape((c+1,r+1))
    return matrix_distance.T
```

```python
def py_to_png(depth_array, output_path):
    z = (65535*((depth_array - depth_array.min())/depth_array.ptp())).astype(np.uint16)
    numpngw.write_png(output_path, z)
```

```python
def read_one_zip(input_zip):
    for j in range(1, 104):
        path_to_data = "home/qiyan/Downloads/"
        relative_path = "Coor("+str(j)+").txt"
        input_path = path_to_data+relative_path 
    
        drone, pixels = txt_to_py(input_zip[input_path].decode("utf-8"))
        
        depth_array = compute_depth(drone,pixels)
        
        output_path = "data\\converted_data\\depth{}-{}.png".format(1, j)
        py_to_png(depth_array,output_path)
```
